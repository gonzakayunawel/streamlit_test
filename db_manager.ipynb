```python
import sqlite3

conn = sqlite3.connect("form_app.db")

def create_users_register_table():
    conn = sqlite3.connect("form_app.db")
    cursor = conn.cursor()
    cursor.execute(
        """
        CREATE TABLE IF NOT EXISTS users_register (
            id INTEGER PRIMARY KEY AUTOINCREMENT,
            name TEXT,
            email TEXT,
            run TEXT,
            curso TEXT,
            electivo_1 TEXT,
            electivo_2 TEXT,
            electivo_3 TEXT,
            electivo_fg TEXT,
            created_at DATETIME DEFAULT CURRENT_TIMESTAMP
        )
        """
    )
    conn.commit()
    conn.close()

def create_email_table():
    conn = sqlite3.connect("form_app.db")
    cursor = conn.cursor()
    cursor.execute(
        """
        CREATE TABLE IF NOT EXISTS emails (
            id INTEGER PRIMARY KEY AUTOINCREMENT,
            email TEXT,
            created_at DATETIME DEFAULT CURRENT_TIMESTAMP
        )
        """
    )
    conn.commit()
    conn.close()
    
def create_electives_table_control():
    conn = sqlite3.connect("form_app.db")
    cursor = conn.cursor()
    cursor.execute(
        """
        CREATE TABLE IF NOT EXISTS electives_control (
            id INTEGER PRIMARY KEY AUTOINCREMENT,
            electivos_disponibles TEXT,
            cantidad_de_inscritos INTEGER,
            cupo_disponible INTEGER,
            cupo_asignado INTEGER,
            electivos_asignados TEXT
        )
        """
    )
    conn.commit()
    conn.close()
```

```python
def insert_email(email):
    conn = sqlite3.connect('form_app.db')
    cursor = conn.cursor()
    cursor.execute(
        """
        INSERT INTO emails (email)
        VALUES (?)
        """,
        (email,),
    )
    conn.commit()
    conn.close()
```

```python
insert_email("micorreo@gmail.com")
insert_email("tucorreo@gmail.com")
```

```python
def verify_email(email):
    conn = sqlite3.connect("form_app.db")
    cursor = conn.cursor()
    cursor.execute("SELECT * FROM emails WHERE email = ?", (email,))
    result = cursor.fetchone()
    if result:
        conn.close()
        return True
        
    else:
        conn.close()
        return False
```

```python
def insert_user(
    name, run, email, curso, electivo_1, electivo_2, electivo_3, electivo_fg
):
    conn = sqlite3.connect("form_app.db")
    cursor = conn.cursor()
    cursor.execute(
        "INSERT INTO users_register (name, run, email, curso, electivo_1, electivo_2, electivo_3, electivo_fg) VALUES (?, ?, ?, ?, ?, ?, ?, ?)",
        (name, run, email, curso, electivo_1, electivo_2, electivo_3, electivo_fg),
    )
    conn.commit()
    conn.close()
    return True
```

```python
def insert_elective(
    nombre_electivo, cantidad_de_inscritos, cupo_disponible, cupo_asignado
):
    conn = sqlite3.connect("form_app.db")
    cursor = conn.cursor()
    cursor.execute(
        "INSERT INTO electives_control (electivos_disponibles, cantidad_de_inscritos, cupo_disponible, cupo_asignado, electivos_asignados) VALUES (?, ?, ?, ?, ?)",
        (
            nombre_electivo,
            cantidad_de_inscritos,
            cupo_disponible,
            cupo_asignado,
        ),
    )
    conn.commit()
    conn.close()
    return True
```

```python
def drop_columns(table_name, columns):
    conn = sqlite3.connect("form_app.db")
    cursor = conn.cursor()
    for column in columns:
        cursor.execute(f"ALTER TABLE {table_name} DROP COLUMN {column}")
    conn.commit()
    conn.close()
    return True
```

```python
drop_columns("electives_control", ["electivos_asignados"])
```
